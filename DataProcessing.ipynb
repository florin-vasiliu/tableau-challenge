```python
import pandas as pd
import numpy as np
from IPython.display import clear_output
import time
```

```python
#Get data
city_bike_df = pd.read_csv("DownloadedData/202005-citibike-tripdata.csv")
```

```python
city_bike_df.set_index("bikeid").loc[18015].iloc[0]
```

```python
#Slice data to get start time of next trip
pilot_city_bike_df = city_bike_df[["bikeid","starttime"]]
pilot_city_bike_df
```

```python
#order data by bike id in chronological order, and remove first row (see next step)
pilot_city_bike_df = pilot_city_bike_df.sort_values(["bikeid", "starttime"]).reset_index()
pilot_city_bike_df = pilot_city_bike_df.drop(0).reset_index().drop(["level_0","index"], axis=1)
pilot_city_bike_df
```

```python
#by ordering in chronological order the bike id both dataframes
#and removing the first row from pilot_city_bike_df, after left joining we get
#the "starttime" from pilot_city_bike_df as "nextstarttime" of city_bike_df
city_bike_df = city_bike_df.sort_values(["bikeid", "starttime"]).reset_index()
city_bike_df = city_bike_df.join(pilot_city_bike_df["starttime"], rsuffix="_r").rename(columns={"starttime_r":"nextstarttime"})
city_bike_df
```

```python
#data clean-up: remove values from last recorded bike id
bikeid_list = city_bike_df["bikeid"].unique()
city_bike_df = city_bike_df.set_index("bikeid").sort_values(["bikeid","starttime"])
for bikeid in bikeid_list:
    print("bikeid", bikeid)
    try: 
        city_bike_df.loc[bikeid].iloc[-1, 15] = "NaN"
    except:
        city_bike_df.loc[bikeid,"nextstarttime"] = "NaN"
    
```
